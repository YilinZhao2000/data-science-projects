```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

iris = load_iris()
iris
```

```python
# todo:2.Specify feature columns and target columns
X = iris.data[:,2].reshape(-1,1) # the data on the 3rd columns
y = iris.data[:,3] # the data on the 4th columns
print(X)
print(y)
```

```python
#to do:line regression
line = LinearRegression()
```

```python
#to do:3.3.Divide the dataset: training set and test set 
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
import random
random.seed(4)
nums=[random.randint(1,100) for i in range(10)]
nums
```

```python
# fitting
line.fit(X_train,y_train)
print("weight: ",line.coef_)
print("intercept: ",line.intercept_)
```

```python
#todo: forecast
Y = line.predict(X_test)
print("forecasting value: ",Y)
print("real value: ",y_test)
```

```python
#todo: visualization
plt.figure(figsize=(15,6))
plt.rcParams['font.family']='SimHei'
plt.plot(y_test,label='real value',color='red',marker='o')
plt.plot(Y,label='forecasting value',color='yellow',marker='D')
plt.xlabel('number of the forecasting dateset')
plt.ylabel('date value')
plt.legend()
plt.show()
```

```python
#todo:Model evaluation
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
print("Mean Squared Error: ",mean_squared_error(y_test,Y))
print("Mean Absolute Error: ",mean_absolute_error(y_test,Y))
print("Training set: ",r2_score(y_train,line.predict(X_train)))
print("Test set: ",r2_score(y_test,Y))
print(line.score(X_train,y_train))
print(line.score(X_test,y_test))
```
